# Część 1: Klasy i Obiekty - Podstawy

## Czym są Klasy i Obiekty?

- **Klasa**: Szablon do tworzenia obiektów (jak foremka do ciastek)
- **Obiekt**: Instancja klasy (jak prawdziwe ciastko)
- **Atrybuty**: Dane przechowywane w obiekcie
- **Metody**: Funkcje należące do obiektu

## Podstawowa Składnia

```python
class ClassName:
    def __init__(self, param1, param2):
        self.attribute1 = param1
        self.attribute2 = param2
    
    def method_name(self):
        return self.attribute1
```

## Ćwiczenie 1.2: Praca z Wieloma Obiektami

Utwórz klasę `Student` z:
- Atrybuty: `name`, `student_id`, `grades` (list)
- Method `add_grade(grade)` dodającą ocenę do listy

````python
# Twój kod tutaj:
class Student:
    def __init__(self, name, student_id):
        self.name = name
        self.student_id = student_id
        self.grades = []

    def add_grade(self, grade):
        self.grades.append(grade)

    def calculate_average(self):
        if len(self.grades) == 0:
            return 0

        return sum(self.grades) / len(self.grades)
````

# Część 2: Hermetyzacja

## What is Hermetyzacja?

Hermetyzacja is the bundling of data and methods, while controlling access to that data.

## Access Modifiers in Python

- **Public**: `attribute` - Accessible from anywhere
- **Protected**: `_attribute` - Intended for internal use (convention)
- **Private**: `__attribute` - Name mangling (harder to access from outside)

## Why Hermetyzacja?

- Protect data integrity
- Control how data is accessed/modified
- Hide implementation details

## Ćwiczenie 2.2: Dekorator Propertys

Utwórz klasę `Temperature` z:
- Private attribute: `__celsius`
- Property `celsius` with getter and setter
- Property `fahrenheit` (calculated from celsius, read-only)
- Validation: celsius must be >= -273.15 (absolute zero)

````python
# Twój kod tutaj:
class Temperature:
    def __init__(self, celsius):
        self.celsius = celsius

    @property
    def celsius(self):
        return self.__celsius

    @celsius.setter
    def celsius(self, value):
        if value >= -273.15:
            self.__celsius = value
        else:
            raise ValueError("Temperature cannot be below absolute zero")

    @property
    def fahrenheit(self):
        return (self.__celsius * 9/5) + 32
````

````python
# Przetestuj swój kod:
temp = Temperature(25)
print(f"Celsius: {temp.celsius}°C")
print(f"Fahrenheit: {temp.fahrenheit}°F")

temp.celsius = 30
print(f"New Celsius: {temp.celsius}°C")
print(f"New Fahrenheit: {temp.fahrenheit}°F")

# Try setting invalid temperature
try:
    temp.celsius = -300
except ValueError as e:
    print(f"Error: {e}")
````

# Część 3: Dziedziczenie

## What is Dziedziczenie?

Dziedziczenie allows a class to inherit attributes and methods from another class.

- **Base/Parent/Superclass**: The class being inherited from
- **Derived/Child/Subclass**: The class that inherits

## Why Use Dziedziczenie?

- Code reuse
- Hierarchical classification
- Polimorfizm support

## Syntax

```python
class Parent:
    pass

class Child(Parent):
    pass
```

## Ćwiczenie 3.1: Basic Dziedziczenie

Create a base class `Vehicle` z:
- Atrybuty: `brand`, `model`, `year`
- Method `start_engine()` that prints "Engine starting..."

Create a child class `Car` that:
- Inherits from `Vehicle`
- Adds attribute: `num_doors`
- Overrides `start_engine()` with a specific message

````python
# Twój kod tutaj:
class Vehicle:
    def __init__(self, brand, model, year):
        self.brand = brand
        self.model = model
        self.year = year

    def start_engine(self):
        print("Engine starting...")


class Car(Vehicle):
    def __init__(self, brand, model, year, num_doors):
        super().__init__(brand, model, year)
        self.num_doors = num_doors

    def start_engine(self):
        print("Car engine starting...")
````

````python
# Przetestuj swój kod:
vehicle = Vehicle("Generic", "Vehicle", 2020)
vehicle.start_engine()

car = Car("Toyota", "Camry", 2023, 4)
car.start_engine()
print(f"This car has {car.num_doors} doors")
````

# Część 4: Polimorfizm

## What is Polimorfizm?

Polimorfizm means "many forms" - the ability of different objects to respond to the same method call in their own way.

## Types of Polimorfizm

1. **Nadpisywanie Metod**: Child class redefines parent's method
2. **Duck Typing**: "If it walks like a duck and quacks like a duck..."
3. **Operator Overloading**: Define custom behavior for operators

## Ćwiczenie 4.1: Nadpisywanie Metod

Create a base class `Shape` with method `area()`

Create child classes that override `area()`:
- `Circle` (radius)
- `Rectangle` (width, height)
- `Triangle` (base, height)

````python
# Twój kod tutaj:
import math

class Shape:
    def area(self):
        pass


class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius ** 2


class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height


class Triangle(Shape):
    def __init__(self, base, height):
        self.base = base
        self.height = height

    def area(self):
        return (self.base * self.height) / 2
````

````python
# Przetestuj swój kod:
shapes = [
    Circle(5),
    Rectangle(4, 6),
    Triangle(3, 8)
]

for shape in shapes:
    print(f"{shape.__class__.__name__}: Area = {shape.area():.2f}")
````

# Część 5: Abstrakcja

## What is Abstrakcja?

Abstrakcja hides complex implementation details and shows only essential features.

## Abstrakcyjne Klasy Bazowe (ABC)

- Cannot be instantiated directly
- Force child classes to implement specific methods
- Use `@abstractmethod` decorator

## Ćwiczenie 5.1: Abstract Base Classes

Create an abstract class `Payment` z:
- Abstract method: `process_payment(amount)`
- Concrete method: `print_receipt(amount)`

Create concrete classes:
- `CreditCardPayment`
- `PayPalPayment`
- `CashPayment`

````python
# Twój kod tutaj:
from abc import ABC, abstractmethod

class Payment(ABC):
    @abstractmethod
    def process_payment(self, amount):
        pass

    def print_receipt(self, amount):
        print(f"Receipt: Payment of {amount}")


class CreditCardPayment(Payment):
    def process_payment(self, amount):
        print(f"Processing credit card payment of {amount}")


class PayPalPayment(Payment):
    def process_payment(self, amount):
        print(f"Processing PayPal payment of {amount}")


class CashPayment(Payment):
    def process_payment(self, amount):
        print(f"Processing cash payment of {amount}")
````

````python
# Przetestuj swój kod:
payments = [
    CreditCardPayment(),
    PayPalPayment(),
    CashPayment()
]

for payment in payments:
    payment.process_payment(100)
    payment.print_receipt(100)
    print()
````

### 💡 Rozwiązanie do Części 5

````python
# Rozwiązanie 5.1: Abstract Payment classes
from abc import ABC, abstractmethod

class PaymentSolution(ABC):
    @abstractmethod
    def process_payment(self, amount):
        pass
    
    def print_receipt(self, amount):
        print(f"Receipt: ${amount} paid via {self.__class__.__name__}")

class CreditCardPaymentSolution(PaymentSolution):
    def process_payment(self, amount):
        print(f"Processing ${amount} via Credit Card...")

class PayPalPaymentSolution(PaymentSolution):
    def process_payment(self, amount):
        print(f"Processing ${amount} via PayPal...")

class CashPaymentSolution(PaymentSolution):
    def process_payment(self, amount):
        print(f"Received ${amount} in cash")
````

## Ćwiczenie 6.2: Operator Overloading

Utwórz klasę `Vector2D` z:
- Atrybuty: `x`, `y`
- Implement `__add__()` for vector addition
- Implement `__sub__()` for vector subtraction
- Implement `__mul__()` for scalar multiplication
- Implement `__eq__()` for equality comparison
- Implement `__str__()` for display

````python
# Twój kod tutaj:
class Vector2D:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Vector2D(self.x + other.x, self.y + other.y)

    def __sub__(self, other):
        return Vector2D(self.x - other.x, self.y - other.y)

    def __mul__(self, scalar):
        return Vector2D(self.x * scalar, self.y * scalar)

    def __eq__(self, other):
        return self.x == other.x and self.y == other.y

    def __str__(self):
        return f"({self.x}, {self.y})"
````

````python
# Przetestuj swój kod:
v1 = Vector2D(1, 2)
v2 = Vector2D(3, 4)

print(f"v1: {v1}")
print(f"v2: {v2}")
print(f"v1 + v2 = {v1 + v2}")
print(f"v2 - v1 = {v2 - v1}")
print(f"v1 * 3 = {v1 * 3}")
print(f"v1 == v2: {v1 == v2}")
print(f"v1 == Vector2D(1, 2): {v1 == Vector2D(1, 2)}")
````

## Ćwiczenie 6.3: Metody Kontenerowe

Utwórz klasę `Playlist` z:

- Implement `__len__()` to return number of songs
- Implement `__getitem__()` to access songs by index

````python
# Twój kod tutaj:
class Playlist:
    def __init__(self, name):
        self.name = name
        self.songs = []

    def add_song(self, song):
        self.songs.append(song)

    def __len__(self):
        return len(self.songs)

    def __getitem__(self, index):
        return self.songs[index]

    def __iter__(self):
        return iter(self.songs)
````

# Część 7: Kompozycja

## What is Kompozycja?

Kompozycja is a "has-a" relationship where one class contains objects of other classes.

## Kompozycja vs Dziedziczenie

- **Dziedziczenie**: "is-a" relationship (Dog is-a Animal)
- **Kompozycja**: "has-a" relationship (Car has-a Engine)

**General rule**: Favor composition over inheritance for flexibility!

## Ćwiczenie 7.1: Relacja "Ma-"

Create a composition relationship:
1. `Engine` class - attributes: `horsepower`, `type`
2. `Car` class - has an `Engine` object
3. Demonstrate composition works differently from inheritance

````python
# Twój kod tutaj:
class Engine:
    def __init__(self, horsepower, type):
        self.horsepower = horsepower
        self.type = type


class Car:
    def __init__(self, brand, model, engine):
        self.brand = brand
        self.model = model
        self.engine = engine

    def display_info(self):
        print(f"{self.brand} {self.model}")
        print(f"Engine: {self.engine.type}, {self.engine.horsepower} HP")
````

````python
# Przetestuj swój kod:
engine = Engine(200, "V6")
car = Car("Toyota", "Camry", engine)
car.display_info()
````

# Część 8: Metody Klasowe i Metody Statyczne

## Types of Methods

1. **Instance Methods**: Regular methods that work with instance data (use `self`)
2. **Class Methods**: Work with class data (use `@classmethod` and `cls`)
3. **Static Methods**: Don't access instance or class data (use `@staticmethod`)

## When to Use Each?

- **Instance method**: When you need to access/modify instance data
- **Class method**: Alternative constructors or when working with class attributes
- **Static method**: Utility functions related to the class but don't need class/instance data

## Ćwiczenie 8.1: Class Methods

Utwórz klasę `Date` z:
- Atrybuty instancji: `year`, `month`, `day`
- Instance method: `display()` to show the date
- Class method: `from_string(date_string)` to create Date from "DD-MM-YYYY"
- Class method: `today()` to create Date with current date

````python
# Twój kod tutaj:
from datetime import datetime

class Date:
    def __init__(self, year, month, day):
        self.year = year
        self.month = month
        self.day = day

    def display(self):
        print(f"{self.day:02d}-{self.month:02d}-{self.year}")

    @classmethod
    def from_string(cls, date_string):
        day, month, year = map(int, date_string.split("-"))
        return cls(year, month, day)

    @classmethod
    def today(cls):
        now = datetime.now()
        return cls(now.year, now.month, now.day)
````

````python
# Przetestuj swój kod:
# Using regular constructor
date1 = Date(2024, 12, 3)
date1.display()

# Using class method from_string
date2 = Date.from_string("25-12-2024")
date2.display()

# Using class method today
date3 = Date.today()
date3.display()
````

## Ćwiczenie 8.2: Static Methods

Utwórz klasę `MathOperations` with static methods:
- `is_prime(number)` - check if number is prime
- `factorial(number)` - calculate factorial
- `fibonacci(n)` - return nth Fibonacci number

````python
# Twój kod tutaj:
class MathOperations:

    @staticmethod
    def is_prime(number):
        if number < 2:
            return False
        for i in range(2, int(number ** 0.5) + 1):
            if number % i == 0:
                return False
        return True

    @staticmethod
    def factorial(number):
        result = 1
        for i in range(1, number + 1):
            result *= i
        return result

    @staticmethod
    def fibonacci(n):
        if n <= 0:
            return 0
        if n == 1:
            return 1

        a, b = 0, 1
        for _ in range(2, n + 1):
            a, b = b, a + b
        return b
````

````python
# Przetestuj swój kod:
print(f"Is 17 prime? {MathOperations.is_prime(17)}")
print(f"Is 20 prime? {MathOperations.is_prime(20)}")
print(f"Factorial of 5: {MathOperations.factorial(5)}")
print(f"10th Fibonacci number: {MathOperations.fibonacci(10)}")
````

## What are Wzorce Projektowe?

Design patterns are reusable solutions to common problems in software design.

## Common Patterns

1. **Singleton**: Ensure only one instance exists
2. **Factory**: Create objects without specifying exact class
3. **Observer**: Objects observe and react to events
4. **Strategy**: Define family of algorithms

## Ćwiczenie 9.1: Wzorzec Singleton

Create a `Logger` class that:
- Only allows one instance to exist (Singleton)
- Has method `log(message)` to log messages
- Stores all logs in a list

````python
# Twój kod tutaj:
class Logger:
    _instance = None

    def __new__(cls):
        if cls._instance is None:
            cls._instance = super(Logger, cls).__new__(cls)
            cls._instance._logs = []
        return cls._instance

    def log(self, message):
        self._logs.append(message)

    def get_logs(self):
        return self._logs
````

````python
# Przetestuj swój kod:
logger1 = Logger()
logger2 = Logger()

print(f"Same instance? {logger1 is logger2}")  # Should be True

logger1.log("First message")
logger2.log("Second message")

print("\nAll logs:")
for log in logger1.get_logs():
    print(f"  {log}")
````

## Ćwiczenie 9.2: Wzorzec Fabryki

Create a `ShapeFactory` that:
- Has a static method `create_shape(shape_type, *args)`
- Returns `Circle`, `Rectangle`, or `Triangle` based on `shape_type`
- Reuse your Shape classes from Exercise 4.1

````python
# Twój kod tutaj:
import math

class Shape:
    def area(self):
        pass


class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius ** 2


class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height


class Triangle(Shape):
    def __init__(self, base, height):
        self.base = base
        self.height = height

    def area(self):
        return (self.base * self.height) / 2


class ShapeFactory:
    @staticmethod
    def create_shape(shape_type, *args):
        if shape_type == "circle":
            return Circle(*args)
        elif shape_type == "rectangle":
            return Rectangle(*args)
        elif shape_type == "triangle":
            return Triangle(*args)
        else:
            raise ValueError("Unknown shape type")

````

````python
# Przetestuj swój kod:
shapes = [
    ShapeFactory.create_shape("circle", 5),
    ShapeFactory.create_shape("rectangle", 4, 6),
    ShapeFactory.create_shape("triangle", 3, 8)
]

for shape in shapes:
    print(f"{shape.__class__.__name__}: Area = {shape.area():.2f}")
````

# Część 10: Projekt Końcowy - System Zarządzania Biblioteką

## Project Requirements

Create a complete library management system with the following classes:

### 1. Book Class
- Atrybuty: `title`, `author`, `isbn`, `is_available`
- Metody: `borrow()`, `return_book()`

### 2. Member Class
- Atrybuty: `name`, `member_id`, `borrowed_books` (list)
- Metody: `borrow_book(book)`, `return_book(book)`, `display_borrowed()`

### 3. Library Class
- Atrybuty: `books` (list), `members` (dictionary)
- Metody:
  - `add_book(book)`
  - `register_member(member)`
  - `lend_book(member_id, isbn)`
  - `receive_book(member_id, isbn)`
  - `display_available_books()`
  - `display_member_info(member_id)`

### Bonus Features (Optional)
- Add due dates for books
- Add fines for late returns
- Search functionality by title/author
- Book categories

````python
# Your implementation here:
class Book:
    def __init__(self, title, author, isbn):
        self.title = title
        self.author = author
        self.isbn = isbn
        self.is_available = True

    def borrow(self):
        if self.is_available == True:
            self.is_available = False

    def return_book(self):
        self.is_available = True


class Member:
    def __init__(self, name, member_id):
        self.name = name
        self.member_id = member_id
        self.borrowed_books = []

    def borrow_book(self, book):
        if book.is_available:
            book.borrow()
            self.borrowed_books.append(book)

    def return_book(self, book):
        if book in self.borrowed_books:
            self.borrowed_books.remove(book)
            book.return_book()

    def display_borrowed(self):
        print("Books borrowed by", self.name)
        for book in self.borrowed_books:
            print("-", book.title)


class Library:
    def __init__(self, name):
        self.name = name
        self.books = []
        self.members = {}

    def add_book(self, book):
        self.books.append(book)

    def register_member(self, member):
        self.members[member.member_id] = member

    def lend_book(self, member_id, isbn):
        member = self.members.get(member_id)

        if member:
            for book in self.books:
                if book.isbn == isbn and book.is_available:
                    member.borrow_book(book)

    def receive_book(self, member_id, isbn):
        member = self.members.get(member_id)

        if member:
            for book in member.borrowed_books:
                if book.isbn == isbn:
                    member.return_book(book)

    def display_available_books(self):
        print("Available books:")
        for book in self.books:
            if book.is_available:
                print("-", book.title, "by", book.author)

    def display_member_info(self, member_id):
        member = self.members.get(member_id)

        if member:
            member.display_borrowed()
````

````python
# Test your library system:
# Create library
library = Library("City Library")

# Add books
book1 = Book("1984", "George Orwell", "ISBN001")
book2 = Book("The Hobbit", "J.R.R. Tolkien", "ISBN002")
book3 = Book("Python Crash Course", "Eric Matthes", "ISBN003")

library.add_book(book1)
library.add_book(book2)
library.add_book(book3)

# Register members
member1 = Member("Alice", "M001")
member2 = Member("Bob", "M002")

library.register_member(member1)
library.register_member(member2)

# Display available books
print("\n=== Available Books ===")
library.display_available_books()

# Borrow books
print("\n=== Borrowing Books ===")
library.lend_book("M001", "ISBN001")
library.lend_book("M001", "ISBN002")
library.lend_book("M002", "ISBN003")

# Display member info
print("\n=== Member Information ===")
library.display_member_info("M001")

# Display available books
print("\n=== Available Books After Borrowing ===")
library.display_available_books()

# Return a book
print("\n=== Returning Books ===")
library.receive_book("M001", "ISBN001")

# Display updated info
print("\n=== Updated Available Books ===")
library.display_available_books()

print("\n=== Updated Member Information ===")
library.display_member_info("M001")
````

### 💡 Pełne Rozwiązanie Projektu Końcowego

````python
# Pełne Rozwiązanie Systemu Zarządzania Biblioteką

class BookSolution:
    def __init__(self, title, author, isbn):
        self.title = title
        self.author = author
        self.isbn = isbn
        self.is_available = True
    
    def borrow(self):
        if self.is_available:
            self.is_available = False
            return True
        return False
    
    def return_book(self):
        self.is_available = True
    
    def __str__(self):
        status = "Available" if self.is_available else "Borrowed"
        return f"'{self.title}' by {self.author} [{status}]"


class MemberSolution:
    def __init__(self, name, member_id):
        self.name = name
        self.member_id = member_id
        self.borrowed_books = []
    
    def borrow_book(self, book):
        if book.borrow():
            self.borrowed_books.append(book)
            print(f"{self.name} borrowed '{book.title}'")
            return True
        else:
            print(f"'{book.title}' is not available")
            return False
    
    def return_book(self, book):
        if book in self.borrowed_books:
            book.return_book()
            self.borrowed_books.remove(book)
            print(f"{self.name} returned '{book.title}'")
            return True
        else:
            print(f"{self.name} doesn't have '{book.title}'")
            return False
    
    def display_borrowed(self):
        if not self.borrowed_books:
            print(f"{self.name} has no borrowed books")
        else:
            print(f"{self.name}'s borrowed books:")
            for book in self.borrowed_books:
                print(f"  - {book.title} by {book.author}")


class LibrarySolution:
    def __init__(self, name):
        self.name = name
        self.books = {}
        self.members = {}
    
    def add_book(self, book):
        self.books[book.isbn] = book
        print(f"Added '{book.title}' to library")
    
    def register_member(self, member):
        self.members[member.member_id] = member
        print(f"Registered member: {member.name} ({member.member_id})")
    
    def lend_book(self, member_id, isbn):
        if member_id not in self.members:
            print(f"Member {member_id} not found")
            return False
        
        if isbn not in self.books:
            print(f"Book {isbn} not found")
            return False
        
        member = self.members[member_id]
        book = self.books[isbn]
        return member.borrow_book(book)
    
    def receive_book(self, member_id, isbn):
        if member_id not in self.members:
            print(f"Member {member_id} not found")
            return False
        
        if isbn not in self.books:
            print(f"Book {isbn} not found")
            return False
        
        member = self.members[member_id]
        book = self.books[isbn]
        return member.return_book(book)
    
    def display_available_books(self):
        available = [book for book in self.books.values() if book.is_available]
        
        if not available:
            print("No books available")
        else:
            print(f"Available books in {self.name}:")
            for book in available:
                print(f"  - {book}")
    
    def display_member_info(self, member_id):
        if member_id in self.members:
            member = self.members[member_id]
            print(f"\nMember: {member.name} ({member.member_id})")
            member.display_borrowed()
        else:
            print(f"Member {member_id} not found")
````
